Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 

```python
a = float(input())
b = float(input())

r = min(a, b) / 2
area = a * b - 3.14 * r * r

print(f"{area:.2f}")
```

Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem

```python
c = input()

if 'A' <= c <= 'Z':
    print(c.lower())
elif 'a' <= c <= 'z':
    print(c.upper())
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
a,b,c = map(int, input().split())
if(a+b>c and a+c>b and b+c>a):
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    print('Dien tich tam giac la: ', round(s,1))
else:
    print('Khong phai la 3 canh cua tam giac')
```

Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh.\
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng\
Input:\
Ten chu ho: Pham Xuan Toai
Chi so thang truoc: 1020
Chi so thang nay: 1070
\
Ouput:\
Ho va ten: Pham Xuan Toai
Tien phai tra la: 107136

```python
name = input()
a = int(input())   # chỉ số tháng trước
b = int(input())   # chỉ số tháng này

c = b - a
result = 0

# Bậc 1
if c > 0:
    kwh = min(c, 50)
    result += kwh * 1984
    c -= kwh

# Bậc 2
if c > 0:
    kwh = min(c, 50)
    result += kwh * 2050
    c -= kwh

# Bậc 3
if c > 0:
    kwh = min(c, 100)
    result += kwh * 2380
    c -= kwh

# Bậc 4
if c > 0:
    kwh = min(c, 100)
    result += kwh * 2998
    c -= kwh

# Bậc 5
if c > 0:
    kwh = min(c, 100)
    result += kwh * 3350
    c -= kwh

# Bậc 6
if c > 0:
    result += c * 3460

# VAT 8%
result = int(result * 1.08)

print(f"Ho va ten: {name}")
print(f"Tien phai tra la: {result}")
```
